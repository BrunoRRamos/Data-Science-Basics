## NumPy Basics

### **Imports:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Url do data set
url = 'https://raw.githubusercontent.com/alura-cursos/numpy/dados/apples_ts.csv'

# Carrega os valores de 'apples_ts.csv' em uma matriz bi-dimensional.
# Ignora os valores de string dos nomes das cidades.
data_set = np.loadtxt(url, delimiter=',', usecols= np.arange(1, 88, 1))
```

### **Visualizando informações sobre o __data_set__**

```python
# Retorna o número de dimenções do array
print(f'Número de dimenções: {data_set.ndim}')

# Retorna a quantidade de dados presentes no data_set
print(f'Número de dados: {data_set.size}')

# Retorna uma tupla que se refere a (num linhas, num colunas)
print(f'Número de linhas x colunas: {data_set.shape}')
```

### **Organizando __data_set__**

Com a realização da operação de matriz transposta podemos organizar os dados de modo que a primeira coluna
se refira aos dados dos meses e as demais **5** colunas restantes representem os dados de cada uma das 
cidades no referente mes da primeira coluna.

```python
# Criando variavel com data_set Transposto
data_set_trasposto = data_set.T

data_set_trasposto
```

### **Separando dados**

```python
# Cria uma lista apenas com os dados da primeira coluna
datas = data_set_trasposto[:,0]

# Cria uma lista apenas com os dados da segunda até a sexta coluna
precos = data_set_trasposto[:, 1:6]
```

### **Gerando uma visualização**

```python
# Gera um grafico datas x preços (por cidade)

Moscow = precos[:, 0]
Kaliningrad = precos[:, 1]
Petersburg = precos[:, 2]
Krasnodar = precos[:, 3]
Ekaterinburg = precos[:, 4]

cidades = [Moscow, Kaliningrad, Petersburg, Krasnodar, Ekaterinburg]

meses = np.arange(1, 88, 1)

for cidade in cidades:
    plt.plot(meses, cidade)
```

### **Estudo por ano**

```python
Moscow_ano1 = Moscow[0:12]
Moscow_ano2 = Moscow[12:24]
Moscow_ano3 = Moscow[24:36]
Moscow_ano4 = Moscow[36:48]

anos = [Moscow_ano1, Moscow_ano2, Moscow_ano3, Moscow_ano4]

for ano in anos:
    plt.plot(np.arange(1, 13, 1), ano)
```
